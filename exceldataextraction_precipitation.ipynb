1. Type this formula in any blank column <b>=IF(A1<>"", A1, IF(Z1<>"", Z1, IF(AA1<>"", AA1, "")))</b>. This will give you column with combine data that you need to cut and paste into columnA.
A1=GageID, Z1=PointID, AA1=DeviceID

2. Type this formula in <b>columnAA =IF(ISNUMBER(SEARCH("TOTAL", A2)), TEXT(A1, "mmmm yyyy"), "")</b>. This will give you name of the month and year from above cell if columnA's cell cotain total. We need to do this because we required total values of the precipitation for ecah year.

3. Now apply filter in data and select total in coulmnA so you will get only totals for each month-year. now copy and paste the data in to new sheet.
4. Cut AA column and paste into column A you will get the extracted total precipitation for each month-year.
5. Match the old gageID with the new gageID and copy-paste the data into one file to combine precipitation data of year 2011-2024.

```python
#check the reason for the skipped file

# File skipping log
skipped_files = []

# Process files
for filename in os.listdir(folder_path):
    # Skip temporary files (those starting with ~$)
    if filename.startswith('~$'):
        continue

    # Check if the file is an Excel file
    if filename.endswith('.xlsx'):
        file_path = os.path.join(folder_path, filename)

        try:
            # Load the Excel file
            df = pd.read_excel(file_path, engine='openpyxl', header=1)

            # Check if at least one ID column exists
            first_column = next((col for col in id_headers if col in df.columns), None)

            if not first_column:
                # Log the reason for skipping
                skipped_files.append((filename, "No valid ID column found."))
                continue

            # Check for numeric columns
            columns_to_keep = [first_column] + [col for col in desired_numeric_headers if col in df.columns]
            if len(columns_to_keep) <= 1:
                # Log the reason for skipping
                skipped_files.append((filename, "No matching numeric columns found."))
                continue

        except Exception as e:
            # Log the reason for skipping due to errors
            skipped_files.append((filename, f"Error: {str(e)}"))

# Log skipped files and reasons
if skipped_files:
    print("Skipped Files:")
    for file, reason in skipped_files:
        print(f"{file}: {reason}")
else:
    print("No files were skipped.")
```
